```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
car_crashes=sns.load_dataset("car_crashes")
```

# Performing Data Visualization

```python
#1. PAIRPLOT
sns.pairplot(car_crashes)
plt.title("Pairplot of car crashes data ")
plt.show
```

Inference: The pairplot shows scatterplots of all numerical variables. You can observe relationships between variables, such as "total" and "speeding," "alcohol" and "total," etc. Some variables may have a positive correlation, while others may not show a clear pattern.

```python
#2. Scatterplot of Alcohol vs. Total Crashes
sns.scatterplot(x="alcohol",y="total",data=car_crashes)
plt.title("Alcohol vs. Total Crashes")
plt.xlabel("Alcohol Consumption")
plt.ylabel("Total Crashes")
plt.show()
```

### Inference: The scatterplot shows a possible positive correlation between alcohol consumption and the total number of crashes. As alcohol consumption increases, the total number of crashes tends to increase.

```python
#Histogram of Speeding
sns.histplot(car_crashes["speeding"], bins=10, kde=True)
plt.title("Distribution of Speeding")
plt.xlabel("Speeding Incidents")
plt.ylabel("Frequency")
plt.show()
```

Inference: The histogram displays the distribution of speeding incidents. Most areas have a low number of speeding incidents, with a peak around 2-4 incidents. The distribution is positively skewed.


```python
#Boxplot of Car Insurance Premiums
sns.boxplot(x="ins_premium",data=car_crashes)
plt.title("Boxplot of Car Insurance premium")
plt.xlabel("Car insurance premmiums")
plt.show()
```

Inference: The boxplot shows the spread of car insurance premiums. It reveals that there are some outliers with significantly higher premiums. The majority of premiums are concentrated in a relatively narrow range.

```python
#Correlation Heatmap
corr_matrix = car_crashes.corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

Inference: The heatmap shows the correlation between all numerical variables. For example, "ins_premium" and "total" have a strong positive correlation, while "speeding" and "total" have a weaker positive correlation.

```python
#Scatterplot of Alcohol vs. Speeding
sns.scatterplot(x="alcohol", y="speeding", data=car_crashes)
plt.title("Alcohol vs. Speeding")
plt.xlabel("Alcohol Consumption")
plt.ylabel("Speeding Incidents")
plt.show()
```

Inference: The scatterplot shows a possible positive correlation between alcohol consumption and speeding incidents. As alcohol consumption increases, speeding incidents may also tend to increase.

```python
#Regression Plot of Not Distracted vs. Insured Premiums
sns.regplot(x="not_distracted", y="ins_premium", data=car_crashes)
plt.title("Not Distracted vs. Insured Premiums")
plt.xlabel("Not Distracted Incidents")
plt.ylabel("Insurance Premiums")
plt.show()
```

Inference: The regression plot suggests that there is no strong linear relationship between the number of not distracted incidents and insurance premiums. The data points are scattered with no clear pattern.

```python
#Histogram of No Previous Accidents
sns.histplot(car_crashes['no_previous'], bins=15, kde=True)
plt.title("Distribution of No Previous Accidents")
plt.xlabel("Number of No Previous Accidents")
plt.ylabel("Frequency")
plt.show()
```

Inference: The histogram shows the distribution of the number of drivers with no previous accidents. It suggests that a significant number of drivers have no previous accidents, with a peak around 5-6 drivers.

```python
#Jointplot of Alcohol vs. Insurance Losses
sns.jointplot(x="alcohol", y="ins_losses", data=car_crashes, kind="reg")
plt.suptitle("Alcohol vs. Insurance Losses")
plt.show()
```

Inference: The jointplot displays the relationship between alcohol consumption and insurance losses. The regression line indicates a possible positive correlation, suggesting that higher alcohol consumption may be associated with higher insurance losses.

```python
# Barplot of Abbreviation Frequencies
sns.barplot(x="abbrev", y="total", data=car_crashes)
plt.title("Average Total Crashes by State Abbreviation")
plt.xlabel("State Abbreviation")
plt.ylabel("Average Total Crashes")
plt.show()
```

Inference: The barplot displays the average total crashes by state abbreviation. It shows that some states have higher average total crashes than others.

```python
#Scatterplot of Insured Premiums vs. Insurance Losses
sns.scatterplot(x="ins_premium", y="ins_losses", data=car_crashes)
plt.title("Insured Premiums vs. Insurance Losses")
plt.xlabel("Insurance Premiums")
plt.ylabel("Insurance Losses")
plt.show()
```

Inference: The scatterplot shows the relationship between insured premiums and insurance losses. There doesn't appear to be a strong linear correlation between the two variables.


```python
#Pairplot of Speeding vs. Not Distracted
sns.pairplot(car_crashes[['speeding', 'not_distracted']])
plt.title("Pairplot of Speeding vs. Not Distracted")
plt.show()
```

Inference: The pairplot visualizes the relationship between speeding incidents and not distracted incidents. It suggests that there may be some positive correlation between the two.
